Generate tfrecord from images and xml labels

```python
%%bash

cd workspace/

# generate train set
python scripts/generate_tfrecord.py \
  -x /content/drive/Shareddrives/dataset_foggy/train/labels/xml \
  -i /content/drive/Shareddrives/dataset_foggy/train/images \
  -l data/label_map.pbtxt \
  -o data/train.tfrecord

# generate validation set
python scripts/generate_tfrecord.py \
  -x /content/drive/Shareddrives/dataset_foggy/val/labels/xml \
  -i /content/drive/Shareddrives/dataset_foggy/val/images \
  -l data/label_map.pbtxt \
  -o data/validation.tfrecord
```
